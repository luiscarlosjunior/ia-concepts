# Exemplo completo de um poss√≠vel algoritmo de busca em largura

```python
from collections import defaultdict
```

```python
class Grafo:
  def __init__(self):
    """
    Define a estrutura a ser utilizada para manter 
    a abstra√ß√£o do grafo
    """
    self.grafo = defaultdict(list)

  def adicionarArestas(self, origem, destino):
    # Cria as rela√ß√µes entre o v√©rtice origem e destino
    self.grafo[origem].append(destino)

  # Fun√ß√£o principal  
  def BuscaLargura(self, raiz):
    # Definir uma lista de v√©rtice(n√≥) visitado
    visitado = [False] * (max(self.grafo) + 1)

    # Criar uma fila para os v√©rtices a serem visitados
    # Abrir o n√≥ raiz
    fila = [] 
    fila.append(raiz)
    # marcar o v√©rtice raiz como visitado
    visitado[raiz] = True

    # Intera√ß√£o entre os v√©rtices que pertencem ao n√≥ raiz
    while fila:
      # Retira o v√©rtice da file e o imprime
      solucao = fila.pop(0)
      print(str(solucao) + " ")

      """
      Interage pelos v√©rtices adjacentes (vizinhos)
      verificando se j√° foram visitados, sen√£o marca-os
      como visitado e depois adiciona na fila para serem
      exibidos e verificar se h√° v√©rtices adjacentes
      """
      for vertice in self.grafo[solucao]:
        if visitado[vertice] == False:
          visitado[vertice] = True
          fila.append(vertice)
```

```python
# Criando uma inst√¢ncia da classe Grafo
g = Grafo()
```

## Desenhando o seguinte Grafo

```python
g.adicionarArestas(1,2)
g.adicionarArestas(1,3)
g.adicionarArestas(1,4)
g.adicionarArestas(2,1)
g.adicionarArestas(2,3)
g.adicionarArestas(3,1)
g.adicionarArestas(3,2)
g.adicionarArestas(3,4)
g.adicionarArestas(4,1)
g.adicionarArestas(4,3)

print("Busca em largura: ")
g.BuscaLargura(1)
```
