# Otimização

Processo procurar o melhor valor possível para uma determinada métrica (ou função objetivo) em um espaço de possíveis soluções. O SciPy é uma biblioteca em Python que fornece ferramentas para realizar tarefas de otimização em uma variedade de problemas, desde otimização de funções matemáticas até ajuste de parâmetros de modelos.

## Tópico em Otimização

### Função Objetivo:
Função que avalia se uma solução é factivel ou não e pode ser matemática, física ou relacionada a algum problema específico. O objetivo é encontrar os valores dos parâmetros de entrada (variáveis de decisão) que otimizam essa função.

### Métodos de Otimização:
**minimize**: Usado para minimização de funções escalares.
**minimize_scalar**: Usado para minimização de funções univariadas.
**linprog**: Usado para otimização linear.
**curve_fit**: Usado para ajuste de curvas a dados experimentais.

### Restrições:
Restrições são casos particulares encontrados em problemas. Por exemplo, o valor de x não deve ultrapassar 100.

### Convergência:
A maioria dos métodos de otimização no SciPy é iterativa e busca convergir para uma solução ótima. É importante verificar se o método convergiu com sucesso e se a solução encontrada é realmente um mínimo ou máximo global.

### Sensibilidade a Condições Iniciais:
Métodos de otimização podem convergir para diferentes soluções dependendo das condições iniciais. Portanto, é essencial escolher as condições iniciais apropriadas, por exemplo, informar parâmetros já conhecidos.

### Tolerâncias e Critérios de Parada:
Muitos algoritmos são iterativos, sendo assim, é importante saber quando parar. Pode ser usado a quantidade de iteração, erro/precisão, tempo ou híbrido.

## Exemplo de otimização de minimização

```python
import numpy as np
from scipy.optimize import minimize

# Definir a função que queremos minimizar
def funcao_objetivo(x):
    return x**2 + 4*x + 4

# Suponha um ponto de partida inicial
chute_inicial = 0.0

# Realizar a otimização
resultado = minimize(funcao_objetivo, chute_inicial)

# Extrair o valor de x que minimiza a função
optimal_x = resultado.x

# Valor mínimo da função
minimized_value = resultado.fun

print(f"O valor de x que minimiza a função é: {optimal_x[0]}")
print(f"O valor mínimo da função é: {minimized_value}")
```

### Exemplo de maximização

```python
import numpy as np
from scipy.optimize import minimize

# Definir a função negativa para a maximização
def negative_objective_function(x):
    return -(x**2 - 4*x)

# Suponha um ponto de partida inicial
initial_guess = 0.0

# Realizar a otimização
resultado = minimize(negative_objective_function, initial_guess)

# Extrair o valor de x que maximiza a função
otimo_x = resultado.x

# Valor máximo da função (negativo do mínimo)
valor_maximizado = -resultado.fun

print(f"O valor de x que maximiza a função é: {otimo_x[0]}")
print(f"O valor máximo da função é: {valor_maximizado}")
```

### Exemplo de maximização com restrição

```python
import numpy as np
from scipy.optimize import minimize

# Definir a função que queremos maximizar (negativo da função original)
def funcao_objetivo(x):
    return -(x*2)

# Definir a restrição: -f(x) = -200
def restricao(x):
    return -funcao_objetivo(x) + 200

# Suponha um ponto de partida inicial
initial_guess = 0.0

# Definir a restrição como uma tupla de dicionários
constraints = ({'type': 'eq', 'fun': restricao})

# Realizar a otimização com a restrição
resultado = minimize(funcao_objetivo, initial_guess, constraints=constraints)

# Extrair o valor de x que maximiza a função
otimo_x = resultado.x

# Valor máximo da função (negativo do resultado da função)
valor_maximo = -resultado.fun

print(f"O valor de x que maximiza a função é: {otimo_x[0]}")
print(f"O valor máximo da função é: {valor_maximo}")
```

### Exemplo de otimização com gráfico

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution

# Definir a função que queremos minimizar
def objective_function(xy):
    x, y = xy
    return x**2 + y**2

# Definir limites para x e y
bounds = [(-5, 5), (-5, 5)]

# Gerar uma grade de pontos para o gráfico
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Realizar a otimização global
result = differential_evolution(objective_function, bounds)

# Extrair os valores de x e y que minimizam a função
optimal_xy = result.x

# Valor mínimo da função
minimized_value = result.fun

# Plotar o gráfico de contorno
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.scatter(optimal_xy[0], optimal_xy[1], color='red', marker='x', label='Mínimo Global')
plt.title('Otimização de Função Bidimensional')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print(f"Os valores de x e y que minimizam a função são: {optimal_xy}")
print(f"O valor mínimo da função é: {minimized_value}")
```

# Distribuição probabilidade

As distribuições de probabilidade desempenham um papel fundamental na estatística e na ciência de dados. Elas nos permitem modelar eventos incertos e entender melhor o comportamento de variáveis aleatórias em uma variedade de cenários do mundo real.
Neste artigo, exploraremos as principais distribuições de probabilidade usando a linguagem de programação Python e a biblioteca Scipy, e apresentaremos exemplos práticos de como aplicá-las em problemas reais.

# Distribuição de Poisson    
amplamente utilizada para modelar a ocorrência de eventos em um período de tempo ou espaço. Vamos ver como podemos usar o Python e o Scipy para calcular probabilidades com esta distribuição.
Exemplo: Modelagem de chegada de clientes a um site em um determinado intervalo de tempo.

```python
import scipy.stats as stats

# Taxa média de chegada de clientes por hora
taxa_media = 5

# Probabilidade de 3 clientes chegarem em uma hora
probabilidade = stats.poisson.pmf(5, taxa_media)
print(f"A probabilidade de 3 clientes chegarem em uma hora é de {probabilidade:.2f}")
```

### Exemplo de poisson com gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

taxa_media = 5
x = np.arange(0, 10)
y = stats.poisson.pmf(x, taxa_media)

plt.bar(x, y)
plt.title('Distribuição de Poisson')
plt.xlabel('Número de clientes')
plt.ylabel('Probabilidade')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parâmetro lambda (taxa média de eventos)
lambda_ = 3.0

# Gerar uma série de valores de Poisson
valores_poisson = poisson.rvs(mu=lambda_, size=1000)

# Calcular a frequência de cada valor
unique, counts = np.unique(valores_poisson, return_counts=True)

# Plotar um histograma dos valores gerados com valores nas barras
plt.bar(unique, counts/len(valores_poisson), alpha=0.6, color='g', label='Dados de Poisson')
plt.xlabel('Número de Eventos')
plt.ylabel('Frequência Relativa')
plt.title(f'Distribuição de Poisson (lambda={lambda_})')

# Adicionar os valores nas barras
for i, count in enumerate(counts):
    plt.text(unique[i], count/len(valores_poisson), f'{count}', ha='center', va='bottom')

plt.legend()
plt.grid(True)

# Calcular e plotar a PMF (função de massa de probabilidade)
x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=lambda_)
plt.vlines(x, 0, pmf, colors='r', lw=2, label='PMF de Poisson')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parâmetro lambda (taxa média de eventos)
lambda_ = 3.0

# Gerar uma série de valores de Poisson
valores_poisson = poisson.rvs(mu=lambda_, size=1000)

# Plotar um histograma dos valores gerados
plt.hist(valores_poisson, bins=20, density=True, alpha=0.6, color='g', label='Dados de Poisson')
plt.xlabel('Número de Eventos')
plt.ylabel('Frequência Relativa')
plt.title(f'Distribuição de Poisson (lambda={lambda_})')
plt.legend()
plt.grid(True)

# Calcular e plotar a PMF (função de massa de probabilidade)
x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=lambda_)
plt.vlines(x, 0, pmf, colors='r', lw=2, label='PMF de Poisson')
plt.legend()

plt.show()
```

# Distribuição Binomial   
A distribuição binomial é útil para modelar experimentos com dois resultados possíveis, como sucesso ou fracasso. Vamos ver como usá-la para resolver um problema prático.
Exemplo: Probabilidade de obter pelo menos 4 caras em 6 lançamentos de uma moeda justa.

```python
# Probabilidade de cara em um lançamento
probabilidade_cara = 0.5

# Probabilidade de pelo menos 5 caras em 6 lançamentos
probabilidade = 1 - stats.binom.cdf(3, 6, probabilidade_cara)
print(f"A probabilidade de pelo menos 4 caras em 6 lançamentos é de {probabilidade:.2f}")
```

```python
n = 6
p = 0.5
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)

plt.bar(x, y)
plt.title('Distribuição Binomial')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidade')
plt.show()
```

# Distribuição de Bernoulli   
A distribuição de Bernoulli é usada para modelar experimentos com dois resultados possíveis, como sucesso ou fracasso. Vamos explorar como calcular um intervalo de confiança para a probabilidade de sucesso usando o Python e a biblioteca Scipy.
Exemplo: Suponha que estamos realizando um experimento em que uma moeda é lançada 20 vezes e queremos calcular um intervalo de confiança de 95% para a probabilidade de obter "cara" em cada lançamento.

```python
from scipy.stats import bernoulli
import numpy as np
import math
from scipy.stats import norm

# Número de lançamentos da moeda
n = 20

# Probabilidade de "cara" em cada lançamento
p = 0.5

# Simulando os lançamentos da moeda
x = bernoulli.rvs(p, size=n)

# Estimativa da probabilidade p_hat
p_hat = np.mean(x)

# Nível de confiança
conf = 0.95

# Cálculo do valor crítico z
z = norm.ppf((1 - conf) / 2, loc=0, scale=1)

# Cálculo do intervalo de confiança
L = p_hat - z * math.sqrt(p_hat * (1 - p_hat) / n)
U = p_hat + z * math.sqrt(p_hat * (1 - p_hat) / n)

# Exibindo resultados
print("Resultados dos lançamentos da moeda:", x)
print("Estimativa da probabilidade (p_hat) =", p_hat)
print("Intervalo de Confiança a 95% =", [L, U])
```

# Distribuição Normal   
A distribuição normal é uma das mais comuns e é frequentemente usada para modelar fenômenos naturais. Vamos demonstrar seu uso em um problema de altura de alunos.
Exemplo: Probabilidade de um aluno ter altura entre 160 cm e 180 cm, com média de 170 cm e desvio padrão de 10 cm.

```python
# Média e desvio padrão da distribuição
media = 170
desvio_padrao = 10

# Probabilidade de altura entre 160 cm e 180 cm
probabilidade = stats.norm.cdf(180, loc=media, scale=desvio_padrao) - stats.norm.cdf(160, loc=media, scale=desvio_padrao)
print(f"A probabilidade de um aluno ter altura entre 160 cm e 180 cm é de {probabilidade*100:.2f}%")
```

```python
media = 170
desvio_padrao = 10
x = np.linspace(media - 3*desvio_padrao, media + 3*desvio_padrao, 100)
y = stats.norm.pdf(x, media, desvio_padrao)

plt.plot(x, y)
plt.title('Distribuição Normal')
plt.xlabel('Altura (cm)')
plt.ylabel('Probabilidade')
plt.show()
```

# Distribuição Exponencial    
A distribuição exponencial é comumente usada para modelar o tempo entre eventos consecutivos. Vamos aplicá-la a um cenário de tempo entre chegadas de mensagens.
Exemplo: Probabilidade de não receber uma mensagem durante um período de duas horas, com uma média de chegada de uma mensagem a cada duas horas.

```python
# Taxa média de chegada de mensagens por hora
taxa_media = 0.5

# Probabilidade de não receber uma mensagem em 2 horas
probabilidade = stats.expon.cdf(2, scale=1/taxa_media)
print(f"A probabilidade de não receber uma mensagem em 2 horas é de {probabilidade:.2f}")
```

```python
taxa_media = 0.5
x = np.linspace(0, 5/taxa_media, 100)
y = stats.expon.pdf(x, scale=1/taxa_media)

plt.plot(x, y)
plt.title('Distribuição Exponencial')
plt.xlabel('Tempo (horas)')
plt.ylabel('Probabilidade')
plt.show()
```

# Distribuição Uniforme    
A distribuição uniforme descreve variáveis aleatórias com igual probabilidade de assumir qualquer valor dentro de um intervalo especificado.
Exemplo: Probabilidade do ônibus chegar em menos de 5 minutos, quando o tempo de chegada segue uma distribuição uniforme entre 0 e 10 minutos.

```python
# Probabilidade do ônibus chegar em menos de 5 minutos
probabilidade = (5 - 0) / (10 - 0)
print(f"A probabilidade do ônibus chegar em menos de 5 minutos é de {probabilidade:.2f}")
```

```python
a = 0
b = 10
x = np.linspace(a, b, 100)
y = stats.uniform.pdf(x, a, b-a)

plt.plot(x, y)
plt.title('Distribuição Uniforme')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Probabilidade')
plt.show()
```
